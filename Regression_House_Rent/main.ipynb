# Encoding the object Dataset

```python
df.describe(include = "object")
```

```python
df.drop(['Area Locality','Floor','Posted On'], axis=1, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
cols = ['Area Type', 'City','Furnishing Status','Tenant Preferred','Point of Contact',]

df[cols] = df[cols].apply(LabelEncoder().fit_transform)
```

```python
plt.figure(figsize = (12,8))
sns.heatmap(df.corr() , annot = True , cmap = "YlGnBu")
```

# train test split

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop('Rent',axis=1).values
y=df['Rent'].values
```

```python
x_train ,x_test ,y_train ,y_test=train_test_split(x,y ,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

# Modeling

# RandomForestRegressor

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
x_train =ss.fit_transform(x_train )
x_test=ss.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
random_forest=RandomForestRegressor()
random_forest.fit(x_train , y_train)
print("Random Forest Training Accuracy:", random_forest.score(x_train , y_train))
print("Random Forest Testing Accuracy:", random_forest.score(x_test , y_test))
```

```python
y_pred = random_forest.predict(x_test)
y_pred

y_pred = random_forest.predict(x_test)
df3 = pd.DataFrame({"Y_test": y_test , "Y_pred" : y_pred})
df3.head(20)
```

```python
plt.figure(figsize= (20,6))

plt.plot(df3[:500])
plt.legend(["Actual" , "Predicted"])
```

```python
from sklearn.metrics import mean_absolute_error , mean_absolute_percentage_error , mean_squared_error

mse = mean_squared_error (y_test , y_pred)
print (mse)

mae = mean_absolute_error (y_test , y_pred)
print (mae)

mape = mean_absolute_percentage_error (y_test , y_pred)
print (mape)
```
